```python
import pandas as pd
import numpy as np

period1 = pd.read_csv('2023_period1.csv')
period2 = pd.read_csv('2023_period2.csv')
```

```python
# Define a function to rank and filter the top 10%
def get_top_10_percent(df):
    # Calculate the number of rows that correspond to the top 10%
    top_10_percent = int(len(df) * 0.10)
    
    # Rank the DataFrame by 'factor' in descending order
    df_sorted = df.sort_values(by='factor', ascending=False)
    
    # Return only the top 10% of the rows
    return df_sorted.head(top_10_percent)

# Apply the function to both dataframes
top_10_percent_period1 = get_top_10_percent(period1)
top_10_percent_period2 = get_top_10_percent(period2)

# Display the result
top_10_percent_period1
```
