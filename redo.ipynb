```python
import pandas as pd
import yfinance as yf


df = pd.read_csv('cleaned_data.csv')


start_date = '2022-06-01'
end_date = '2023-12-31'


stock_prices_df = pd.DataFrame()


for symbol in df['Symbol']:
    
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    
    
    adj_close_prices = stock_data['Adj Close'].reset_index()
    
    
    adj_close_prices['Date'] = adj_close_prices['Date'].dt.strftime('%Y-%m-%d')
    
    
    adj_close_prices_pivot = adj_close_prices.pivot_table(index=None, columns='Date', values='Adj Close')
    
    
    adj_close_prices_pivot['Symbol'] = symbol
    
    
    stock_prices_df = pd.concat([stock_prices_df, adj_close_prices_pivot], ignore_index=True)


final_df = pd.merge(df, stock_prices_df, on='Symbol', how='left')


final_df.to_csv('cleaned_data_with_prices2.csv', index=False)

print("Stock prices added successfully!")
```
